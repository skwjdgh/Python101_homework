문자열을 입력받아 팰린드롬인지 판별하라.​
*팰린드롬이란 거꾸로 읽어도 똑같은 문장 또는 단어​

Ex) ‘기러기’, ’mom’ 등

조건​

대소문자 구분 X​
공백 무시

```python
def is_palindrome(text_string,state):
    is_check = text_string.upper().replace(" ", "") # (lower도 상관없음, swapcase() 대소문자 변경, capitalize() 첫 문자만 대문자로) 대소문자 구분 없애기,공백제거
    if is_check == is_check[::-1]: # 역순으로 만들어줌
        state = 1
    return state
```

```python
text_string = input("문자열을 입력하세요.")
state = 0
state = is_palindrome(text_string,state)

print("팰린드롬" if state == 1 else "팰린드롬 아님")
```
