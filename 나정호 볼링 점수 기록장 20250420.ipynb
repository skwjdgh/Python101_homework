# 문제: "볼링 점수 기록장"
설명:
사용자가 이름을 입력하면, 해당 이름으로 텍스트 파일을 생성합니다.
사용자는 10개의 점수를 입력합니다 (각 점수는 0~10 사이 정수).
총점을 계산하여 파일에 저장합니다.
프로그램 종료 후, 파일을 열어 사용자 점수와 총점을 확인할 수 있습니다.

입력 예:
이름을 입력하세요: 홍길동
1번째 프레임 점수 입력: 7
2번째 프레임 점수 입력: 2
...
10번째 프레임 점수 입력: 6

저장될 파일 내용 (홍길동.txt):
홍길동의 볼링 점수
--------------------
1프레임: 7
2프레임: 2
...
10프레임: 6

총점: 78

힌트:

1)input() 함수로 입력 받기

2)리스트에 점수 저장

3)sum()으로 총점 계산

4)open()으로 파일 저장

```python
# 파일입출력 함수
import os
```

```python
# 상수값 선언
default_frame = 10
default_pin = 10
point=[]
```

```python
# 플레이어 이름 입력 및 텍스트 파일 생성
player_name = input("이름을 입력하세요:")
os.makedirs(f"{player_name}",exist_ok =True)
file = f"{player_name}/{player_name}.txt"
```

```python
# 볼링 게임 점수 입력
for is_frame in range(default_frame):
    point.append(int(input(f"{str(is_frame+1)}번째 프레임 점수 입력: ")))
```

```python
# 플레이어 이름 파일에 데이터값 입력
with open(file,"w",encoding ="utf-8") as f:
    f.write(f"{player_name}의 볼링 점수\n")
    for i in range(default_frame):
        f.write(f"{str(i+1)}프레임 : {point[i]}\n")
    f.write(f"총점은 {sum(point)}점 입니다.\n")
```
