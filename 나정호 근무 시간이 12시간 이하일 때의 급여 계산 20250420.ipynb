```python
# 주당 근무시간 기반
def regular_pay(hours, wage):
    return hours * wage

def overtime_pay(hours, wage, standard_hours=12, overtime_rate=1.3):
    overtime_hours = hours - standard_hours
    return (standard_hours * wage) + (overtime_hours * wage * overtime_rate)

pay_calc = {
    True: regular_pay,
    False: overtime_pay
}

hours = float(input("주당 총 근무 시간: "))
wage = float(input("시간당 급여: "))

total_pay = pay_calc[hours <= 12](hours, wage)  # hours <= 12는 조건문 , 조건에 따라 함수를 호출하고 거기에 대한 매개변수를 대입하는 구문
print(f"1주일 급여는 {total_pay:.2f}원 입니다.")
```

```python
# 일당 근무시간 기반

work_hour = int(input("일 근무 시간을 입력하세요."))

# 기본 근무시간
# 9시간
# 기본 시급

min_payment = 10030
default_work_hour = 12

default_payment = min_payment * default_work_hour

overtime_payment = (min_payment * 1.3) * (work_hour - default_work_hour)

daily_payment = default_payment + overtime_payment

print(daily_payment)

#시간외 수당 = (통상임금 * 1.5) * 초과 근무시간.
#야간 수당 = (통상임금 * 1.5) * 야간 근무시간.
#휴일 수당 = (통상임금 * 1.5) * 휴일 근무시간
```
