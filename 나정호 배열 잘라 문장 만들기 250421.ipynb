```python
# 주어진 문자열과 리스트
words = ["apocalypse","meteorology","architect","cheetah"]
slices = [[(1, 2), (5, 6)],[(2, 3)],[(2, 2)],[(1, 1)]]
```

```python
# slices 배열에 따라 문자열을 자르는 함수
def cut_array(words, slices):
    result = [] #자른 결과물을 모아둘 자료형
    for i in range(len(words)):
        is_word = words[i]
        is_slice = slices[i]
        piece ="" # 자른 결과물을 놓을 자료형
        for start, end in is_slice: #리스트 내의 튜플 자료형에 따라 start값과 end값 추출
            piece += is_word[start : end+1] # 배열의 마지막 값은 포함되지 않으므로
        result.append(piece) #루프에서 나온 값을 result에 넣고 루프 반복
    return result
```

```python
#코드

str_array = cut_array(words,slices)

answer = ""
for j in str_array: # 리스트의 요소를 꺼내어 문자열에 집어넣음
    answer += j

print(answer)
```
