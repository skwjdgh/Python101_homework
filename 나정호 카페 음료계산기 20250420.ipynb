1. menu: 딕셔너리 형태로 구성된 음료 이름-가격 쌍 ​

(예: {"아메리카노": 1500, ​"카페라떼": 1800, "초코": 2000})​

입력은 "끝"이라는 문자열이 들어올 때까지 계속됩니다.​

입력되는 모든 문자열은 공백이 없는 한글로 구성됩니다.

2. orders: 문자열 리스트 형태로, 주문한 음료의 이름이 담겨 있음 ​

(예: ["아메리카노", "초코", ​"초코"])​

"총 결제 금액은 N 원입니다." 형태로 출력​ (N은 주문한 메뉴들의 가격 총합)​

단, 메뉴에 없는 입력은 "메뉴에 없습니다. 다시 입력해주세요."를 출력한 뒤 입력을 다시 받습니다.

3. 조건 정리

1) orders에 있는 음료는 menu에 존재함​

2) 음료 이름은 한글 문자열, 공백없음​

3) orders 길이는 1이상 100이하​

```python
max_length = 100
```

```python
# 메뉴표
menu = {"아메리카노": 1500, "​카페라떼": 1800, "초코": 2000, "민트초코" :5000, "카라멜 마키아또":7000}
```

```python
# 주문서
order = []

for i in range(max_length):
    input_order = input("메뉴를 입력하세요.(종료하시려면 종료라고 입력하세요.): ")
    if input_order == "종료":
        break
    if input_order in menu:
        order.append(input_order)
    else:
        print("메뉴에 없습니다. 다시 입력해주세요.")
```

```python
# 결제금액 계산
total_price = 0
total_price = sum(menu[input_order] for input_order in order)
print(f"총 결제 금액은 {total_price} 원입니다.")
```
