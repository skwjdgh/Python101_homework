(사람 대 사람으로 하는 경우)​

1. 문제 출제자는 세 자리의 자연수를 메모합니다. ( 000 ~ 999 의 숫자)​

2. 문제를 맞추는 사람이 세 자리의 수를 얘기합니다.​

3. 숫자만 맞추고 자릿수는 맞지 않는 경우 Ball, 숫자와 자릿수까지 맞으면 Strike,​

  숫자와 자릿수까지 다 틀린 경우 Out이라고 출제자가 말해줍니다.​
 (예를 들어 출제자가 생각해 낸 숫자는 281, 맞추는 사람이 160 이라고 하면 1은 맞지만 자리가 안 맞기 때문에 1ball, 맞추는 사람이 360 이라고 하면 숫자도 자릿수도 안 맞기에 out, 216이라고 하면 1개는 숫자만 맞고 1개는 자릿수까지 맞기 때문에 1ball 1strike)



```python
import random
```

```python
# 숫자를 비교하는 함수
def compare_numbers(number, is_input):
    s_count = 0 # 자릿수 길이를 통해 각 문자열의 위치가 같다면 스트라이크 카운트를
    b_count = 0 # 자릿수 길이를 통해 해당 문자가 저장된 문자열의 일부라면 볼 카운트를

    for i in range(len(number)):
        if is_input[i] == number[i]:
            s_count += 1
        elif is_input[i] in number:
            b_count += 1

    return s_count, b_count
```

```python
# 자연수 입력
number = f"{random.randint(0, 999):03}" #f"{pi:.2f}" : 소수점 두자리,f"{large_number:,}" : 천 단위 구분 등, 다양한 형식이 존재함
is_input = input("세 자리 자연수를 입력하세요.")
```

```python
#스트라이크 볼 확인
s, b = compare_numbers(number, is_input)
if s==0 and b == 0:
    print("Out")
else:
    print(f"{s}스트라이크, {b}볼")
    print(number)
```
