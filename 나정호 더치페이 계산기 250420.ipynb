```python
import random
```

```python
# 인원 및 장소
people=['kim','go','jang','back']
place = ['restaurant','bar','cafe']
```

```python
# 1. 랜덤함수를 통해 인원별 방문 장소 표기

participant = {visited : random.choice([0,1]) for visited in people}
print(participant)
print(place)
```

```python
# 2. 상태에 따른 방문 리스트 생성 0: 식당, 바 1: 바, 카페

visit = {0:['restaurant', 'bar'],1:['bar', 'cafe']}
print(place)
```

```python
# 3. 사람을 장소에 넣기

in_place = {POP : [] for POP in place} #이름들이 들어갈 장소 딕셔너리를 먼저 만들어놓음

for person in people: # 리스트 내의 사람들
    is_place = visit[participant[person]] # 이름에 따라 장소 할당, 즉 participant의 key 값과, visit의 value값을 매칭시킴
    for places in is_place: #
        in_place[places].append(person) #사람에 따른 value값을 가져와 장소 딕셔너리에 해당 사람 이름을 집어넣음

print(in_place)
print(place)
```

```python
# 4. 장소별 총액을 입력 받음
cost = {}
for cost_place in place:
    cost[cost_place] = int(input(f"{cost_place}에서의 총액:"))
print(cost)
```

```python
# 5. 더치페이 계산
dutch_pay = {person: 0 for person in people}

for dutch_place in place:
    if len(in_place[dutch_place]) > 0: #사람이 있을 때만 동작
        dutch = cost[dutch_place]/ len(in_place[dutch_place]) # 위치에 따른 분배금액
        for dutch_person in in_place[dutch_place]: # 해당 위치에 있는 사람에게 분배금액 할당
            dutch_pay[dutch_person] += dutch

print(dutch_pay)
```
