마트에서 장을 볼 때, 어떤 상품은 할인이 적용되고 어떤 상품은 그렇지 않습니다.​

이 문제에서는 할인 조건에 따라 지불해야 할 총 금액을 계산하는 프로그램을 만들어봅니다!

문제 설명​

​

- 물건 가격이 10,000원 이상이면 10% 할인이 적용됩니다.​

- 물건 가격이 10,000원 미만이면 할인이 적용되지 않습니다.​

- 여러 개의 물건을 샀을 때 최종 지불 금액을 계산하세요.

```python
costs = [12000, 8000, 15000, 5000]
```

```python
# 인덱싱을 붙여 조건에 따라 각각의 숫자에 할인값 적용
def is_discount(cost):
    for i,price in enumerate(cost):
        if price >= 10000:
            cost[i] *= 0.9
```

```python
is_discount(costs)
print(f"총 금액은 {int(sum(costs))}원 입니다. ")
```
