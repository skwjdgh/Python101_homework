 강도 평가는 여러 조건을 기준으로 점수 부여, 점수에 따라 강함, 보통, 약함 부여

1. 비밀번호 8자 이상숫 2점
2. 숫자를 하나 이상 포함하면 2점
3. 대문자 포함하면 1점, 소문자 포함1 점
4. 각각 별도의 함수로
5. 점수 계산
    총점 6점
    ,4~5점
    ,3점 이하

```python
# 1. 길이가 8자 이상일 경우, 2점
def length(input_password,point):
    if len(input_password) >= 8:
        point[0] = 2
    return point
```

```python
# 2. 숫자가 하나 이상일 경우, 2점
def num_count(input_password,point):
    count = sum(is_digit.isdigit() for is_digit in input_password)
    if count >= 1:
        point[1] = 2
    return point
```

```python
# 3. 대문자 포함시, 1점
def upper_count(input_password,point):
    upper_count = sum(char.isupper() for char in input_password)
    if upper_count >= 1:
        point[2] = 1
    return point
```

```python
# 4. 소문자 포함시, 1점
def lower_count(input_password,point):
    lower_count = sum(char.islower() for char in input_password)
    if lower_count >= 1:
        point[3] = 1
    return point
```

```python
# 비밀번호 및 점수를 저장할 자료형 선언
input_password = input("비밀번호를 입력하세요")
point = [0,0,0,0]
```

```python
# 비밀번호 강도 검사
length(input_password,point)
num_count(input_password,point)
upper_count(input_password,point)
lower_count(input_password,point)
total = sum(point)

if total >= 6:
    print("strong")
elif total < 6 and total >=4:
    print("normal")
elif total < 4:
    print("week")
```
