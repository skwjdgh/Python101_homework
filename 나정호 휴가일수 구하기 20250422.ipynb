```python
def 휴가일수 구하기(출근일수):
    return 출근일수 // 20


이름과 휴가 ={}
for _ in range(10):
    이름입력 = input("이름을 입력하세요 (그만하려면 '종료' 입력):")
    if 이름입력 =="종료":
        break

    출근일수입력 = int(input(f" {이름입력}의 출근 일 수를 입력하세요:"))
    휴가일수 =휴가일수 계산기 (출근일수입력)

    이름과 휴가[이름입력] = 휴가일수


while True:
    검색이름 = input("\n 검색할 이름을 입력하세요 (종료하려면 '종료' 입력):")
    if 검색이름 == "종료":
        print("검색을 종료합니다.")
        break
    if 검색이름 in 이름과 휴가:
        print(f"{검색이름}의 휴가 일수는 {이름과휴가[검색이름]}일입니다.")
    else:
        print(f"{검색이름}은(는) 목록에 없습니다.")
```

```python
# 휴일 계산함수?
def get_vacation(work_day):
    return work_day//20 #20일 출근시 1일 휴가
```

```python
# 이름을 최대 10개까지 입력받아

name_day_off = {}
for i in range(10): # 글자수 10글자 제한
    input_name = input("이름을 입력하세요 (그만하려면 '종료' 입력):")
    if input_name == "종료":
        break
    work_day = int(input(f" {input_name}의 출근 일 수를 입력하세요:"))
    day_off = get_vacation(work_day)
    name_day_off[input_name] = day_off
```

```python
#  입력된 딕셔너리 자료형에서 해당 이름을 찾음
while True:
    index_name = input("\n 검색할 이름을 입력하세요 (종료하려면 '종료' 입력):")
    if index_name == "종료":
        print("검색을 종료합니다.")
        break
    if index_name in name_day_off:
        print(f"{index_name}의 휴가 일수는 {name_day_off[index_name]}일입니다.")
    else:
        print(f"{index_name}은(는) 목록에 없습니다.")
```
