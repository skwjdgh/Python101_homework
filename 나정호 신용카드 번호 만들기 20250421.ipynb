신용카드 번호 만들기

다음 조건에 맡게 카드 번호를 생성하는 문제

유효한 신용카드 번호 조건

(1)숫자 4, 5 또는 6으로 시작해야 한다.

(2)총 16자리 숫자여야 한다.

(3)숫자(0~9)로만 구성되어야 한다.

(4)4자리씩 하이픈(-)으로 나눌 수는 있지만, 하이픈 외 다른 구분기호는 사용할 수 없다.
예: 1234-5678-9012-3456 (가능)
예: 1234 5678 9012 3456, 1234_5678_9012_3456 (불가능)

(5)같은 숫자가 4번 이상 연속해서 반복되면 안 된다.
예: 4444-4444-4444-4444 (불가능)
예: 4443-4444-4444-4444 (가능)

입력 형식
(1)첫 번째 줄: 입력 받을 카드 번호의 개수 N (1 이상)
(2)다음 N줄: 각각 신용카드 번호가 주어진다.

출력 형식
각 줄마다 해당 카드 번호가 유효하면 "Valid"를,
유효하지 않으면 "Invalid"를 출력하시오.

```python
def valid_number(card_numbers):
    number = card_numbers.replace("-","")
    # 슬라이싱을 통해 숫자 반복 확인
    first_piece = number[0:4]
    second_piece = number[4:8]
    third_piece = number[8:12]
    fourth_piece = number[12:16]

    # 4,5,6으로 시작
    if not (number[0] in ["4","5","6"]):
        return False

    # 숫자 및 16자리 판단
    if len(number) != 16 or not number.isdigit():
        return False

    # '-' 외의 기호 존재 여부
    for check_sign in number:
        if not check_sign.isdigit() and check_sign !='-':
            return False

    # 같은 숫자의 반복 여부
    if first_piece == second_piece == third_piece == fourth_piece:
        return False

    return True
```

```python
n = int(input("검증할 카드 개수를 입력하세요: "))

for card_numbers in range(n):
    card = input("카드 번호를 입력하세요: ")

    if valid_number(card):
        print("Valid")
    else:
        print("Invalid")
```
